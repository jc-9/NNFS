```python
"""
This is a single neruon with 3 inputs into the neuron and a single output.
"""
inputs = [1,2,3]
weights = [0.2, 0.8, -0.5]
bias = 2

output = (inputs[0]*weights[0] + inputs[1]*weights[1] + inputs[2]*weights[2] + bias)
print("Output:",output)
```

4 Inputs and 3 output layer example


```python
"""
Output from a single layer with 3 neurons (1 output per neuron = 3 output values).
Each neuron gets 4 inputs, this is a FULLY CONNECTED LAYER. Not all layers will be fully connected in
real applictions.

Some applicaions will have partically connected layers or layers that "Toggle" on and off during training to help prevent unwanted behaviors. More on this later.
"""


inputs = [1,2,3,2.5]

weights1 = [0.2,0.8,-0.51,1]
weights2 = [0.5,-0.91,0.25,-0.5]
weights3 = [-0.26, -0.27, 0.17,0.87]

bias1 = 2
bias2 = 3
bias3 = 0.5

outputs = [
    # Neuron 1:
    inputs[0]*weights1[0] +
    inputs[1]*weights1[1] +
    inputs[2]*weights1[2] +
    inputs[3]*weights1[3] + bias1,
    # Neuron 2:
    inputs[0]*weights2[0] +
    inputs[1]*weights2[1] +
    inputs[2]*weights2[2] +
    inputs[3]*weights2[3] + bias2,
    # Neuron 3:
    inputs[0]*weights3[0] +
    inputs[1]*weights3[1] +
    inputs[2]*weights3[2] +
    inputs[3]*weights3[3] + bias3]
print(outputs)
```

Take the same code from above and then tranform it into a loop so that we dont have to hardcode it.
This is important to understand as it shortens the code.

```python
"""
Output = input*weight+bias
"""
inputs = [1,2,3,2.5]

# Single list of weights
weights = [
    [0.2,0.8,-0.51,1], #weights1
    [0.5,-0.91,0.25,-0.5], #weights2
    [-0.26, -0.27, 0.17,0.87] #weights3
]

# Single list of bias
bias = [2,3,0.5]

# loop

layer_outputs = [] # This is where we will store the output values. We should expect 3 values in this list

for weights, bias in zip(weights, bias):

    single_neuron_output_without_bias = 0 # Zero this out BC we will reuse this variable to store the output
    single_neuron_output_plus_bias = 0

    for index, (i_input, j_weight) in enumerate(zip(inputs, weights)):

        single_neuron_output_without_bias += i_input*j_weight
        if index < (len(inputs)-1):
            print(f'i_input:{i_input}* weights{j_weight} = {single_neuron_output_without_bias} <- Add this to the sum')

    """
    This is the equivalent for each neuron
    inputs[0]*weights1[0] + inputs[1]*weights1[1] + inputs[2]*weights1[2] + inputs[3]*weights1[3]
    """
    print(' ________________')
    print(f'Single Neuron Output Sum (Input* Weight):{single_neuron_output_without_bias} <- Final Sum')
    print(" ")

    print('Now Add in the Bias')
    single_neuron_output_plus_bias += single_neuron_output_without_bias + bias
    print(f"bias = {bias}")
    print(f'Single Neuron Output Sum (Input* Weight): ({single_neuron_output_without_bias}) + bias ({bias}) : {single_neuron_output_plus_bias} <- Final Sum + Bias')
    print(" ")

    print('Add this to the list')
    layer_outputs.append(single_neuron_output_plus_bias)
    print(f"List Current Value:{layer_outputs}")
    print(" ")


    # single_neuron_output += bias
```

Now do the same thing but using the dot product with numpy

```python
import numpy as np

inputs = [1,2,3,2.5]

# Single list of weights
weights = [
    [0.2,0.8,-0.51,1], #weights1
    [0.5,-0.91,0.25,-0.5], #weights2
    [-0.26, -0.27, 0.17,0.87] #weights3
]

# Single list of bias
bias = [2,3,0.5]

layer_outputs_withnumpy = np.dot(weights, inputs) + bias
print(f'layer outputs using the np.dot method               {layer_outputs_withnumpy}')
print(f'layer_outputs from above using the manual method:   {layer_outputs}')
```
